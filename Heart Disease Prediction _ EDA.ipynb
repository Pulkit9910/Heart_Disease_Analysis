```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("dataset.csv")
```

```python
df.head()
```

# Distribution plot for Numeric column

```python
df['age']
```

```python
sns.distplot(df['age'] , kde = True ,hist = True , rug = True)   # distribution plot of age
                                                                # kde adds a wavy line
```

```python
sns.displot(df['age'] , kde = True , color = 'blue')
```

# Categorical Columns (Pie Charts)

```python
sns.displot(df['trestbps'] , kde = True , color = 'blue')
```

```python
sns.displot(df['chol'] , kde = True , color = 'blue')
```

```python
sns.displot(df['fbs'] , kde = True , color = 'blue')
```

## Pie Charts for Categorical

```python
df.groupby('sex').size().plot(kind = 'pie' , autopct = '%0.2f')    # 0 - Female : 1 - Male
```

```python
df.groupby('cp').size().plot(kind = 'pie' , autopct = '%0.1f')

# Here in pichart 0,1,2,3 represents levels of chest pains
```

```python
df.groupby('restecg').size().plot(kind = 'pie' , autopct = '%0.1f')
```

## ViolinPlot

```python
sns.violinplot(df['sex'])
```

```python
sns.violinplot(df['age'])
```

```python
sns.violinplot(x = df['sex'] , y = df['oldpeak'])
```

```python
sns.violinplot(x = df['age'] , y = df['trestbps'])
```

## Correlation (HeatMap)

```python
df.head()
```

```python
df.corr()              # gives correlation between different columns
                       # The correlation values can fall between -1 and +1. 
                       # If the two variables tend to increase and decrease together,
                       # the correlation value is positive. If one variable increases while the other 
                       # variable decreases, the correlation value is negative.
```

```python
sns.heatmap(df.corr())
```
